```python
#1 loading libraries and dataset.
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
# Load data
df = load_breast_cancer()
X = pd.DataFrame(df.data, columns=df.feature_names)
y = pd.Series(df.target)
X
     
```

```python
# Check missing values
print(X.isnull().sum().sum())
#Null Values is Zero so no need of any cleaning.
```

```python
#Feature-Scaling.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
#Splitting to train-test
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.25,random_state=0)
```

```python
#Implimenting the Algorithms

#LR

#LR works on Sigmoid function and works on binary classfication based on a threshold for example if threshold=0.5,
#then values above it will be classified as 1 and values below it will be 0
#Logistic-Regression is suitable for binary classifications like these
from sklearn.linear_model import LogisticRegression

Logistic=LogisticRegression()
#fitting the model
Logistic.fit(X_train,y_train)
```

```python
#predicting based on x_test
y_logistic_pred=Logistic.predict(X_test)
     
```

```python
from sklearn.ensemble import RandomForestClassifier
#Random-Forest isan ensemble of Decession-tree,as the name suggests its like a forest with multiple decession-trees,Random forest help in reduccing overfitting.
Random_Forest=RandomForestClassifier()

Random_Forest.fit(X_train,y_train)
```

```python
#SVM works by  ]Finding the optimal hyperplane that separates classes with maximum margin.it will be using kernels for non-linear boundaries.
from sklearn.svm import SVC

Support_vector=SVC(kernel='rbf')

Support_vector.fit(X_train,y_train)
     
```

```python
#k-Nearest Neighbors (k-NN) works by Classifing a point based on the majority label among its k nearest neighbors.
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train, y_train)
```

```python
#Using-Accuracy,Precision,Recall and F1 for model-evaluvation.

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

models={
    "Logistic-Regression":Logistic,
    "Random-Forest":Random_Forest,
    "Support-Vector":Support_vector,
    "KNN":knn

}
results=[]
for name,model in models.items():
  y_pred=model.predict(X_test)
  results.append({
      "model":name,
      "accuracy":accuracy_score(y_test,y_pred),
      "Precession":precision_score(y_test,y_pred),
      "Recall":recall_score(y_test,y_pred),
      "F1-score":f1_score(y_test,y_pred)
      })

results
```

```python
#Turning to df for better understanding
results_df=pd.DataFrame(results)
results_df
```
